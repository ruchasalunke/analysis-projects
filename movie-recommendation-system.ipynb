```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
#merging the 2 csv files based on title so it is easy to handle
movies = movies.merge(credits,on='title')
```

```python
movies
```

```python
#extracting important columns
#genres
#keywords
#overview
#title
#movie_id
#cast
#crew
movies = movies[['movie_id','title','genres','overview','keywords','cast','crew']]
```

```python
movies
```

```python
#checking for null values
movies.isnull().sum()
```

```python
#removing null values
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
#checking for duplicates
movies.duplicated().sum()
```

```python
#extracting the genre column
movies.iloc[0].genres
```

```python
#now we just want to extract the value in the name column above, so we make a separate function to extract the values
import ast
def convert(obj):
    l = [] #creating a list
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l    
```

```python
#applying the convert function to genre
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies
```

```python
#applying the convert function to keywords
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies
```

```python
#we will apply the same thing to the cast but we will only consider the first three actors
def convert_cast(obj):
    count = 0
    l = [] #creating a list
    for i in ast.literal_eval(obj):
        if count != 3:
            l.append(i['name'])
            count += 1
        else:
            break
    return l    
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies
```

```python
movies.iloc[0].crew
```

```python
def fetch_director(obj):
    l = [] #creating a list
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l    
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies
```

```python
#now we convert overview column to a list to append to the tags
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies
```

```python
#remove the spaces
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies
```

```python
final_movies = movies[['movie_id','title','tags']]
```

```python
final_movies['tags'] = final_movies['tags'].apply(lambda x: " ".join(x))
```

```python
final_movies['tags'][0]
```

```python
#convert to lower case
final_movies['tags'] = final_movies['tags'].apply(lambda x: x.lower())
```

```python
final_movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(final_movies['tags']).toarray()
```

```python
vectors
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#helper function
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
final_movies['tags'] = final_movies['tags'].apply(stem)
```

```python
final_movies.head()
```

```python
cv.get_feature_names_out()
```

```python
#we calculate cosine distance between the two vectors
#as this is a 5D movie, euclidean distance fails
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(similarity[0])
```

```python
#main function to recommend movie
def recommend(movie):
    movie_index = final_movies[final_movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(final_movies.iloc[i[0]].title)
```

```python
recommend('titanic')
```
